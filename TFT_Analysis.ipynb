```python
import requests
import urllib
import json
import datetime
import time
import pandas as pd
from tqdm import tqdm
```

```python
my_key = ''
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36",
    "Accept-Language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7",
    "Accept-Charset": "application/x-www-form-urlencoded; charset=UTF-8",
    "Origin": "https://developer.riotgames.com",
    "X-Riot-Token": my_key
}
```

```python
today = datetime.datetime.today()
year = today.year
month = today.month
day = today.day
hour = today.hour
```

```python
chal_request_url = f'https://kr.api.riotgames.com/tft/league/v1/challenger?api_key{my_key}'
chal_res = urllib.request.Request(chal_request_url)

for key, value in headers.items():
    chal_res.add_header(key,value)

response = urllib.request.urlopen(chal_res)
result = response.readlines()

result = json.loads(result[0].decode('utf-8'))

if response.getcode() != 200:
    print('Error')

tier = result.get('tier')
entries = result.get('entries')

kr_challenger_username_list = []
kr_challenger_userrank_list = []
kr_challenger_userwins_list = []
kr_challenger_userlosses_list = []
kr_challenger_summonerId_list = []

for player in entries:
    user_name = player.get('summonerName')
    kr_challenger_username_list.append(user_name)
    user_rank = player.get('rank')
    kr_challenger_userrank_list.append(f'{tier} {user_rank}')
    user_wins = player.get('wins')
    kr_challenger_userwins_list.append(user_wins)
    user_losses = player.get('losses')
    kr_challenger_userlosses_list.append(user_losses)
    summonerId = player.get('summonerId')
    kr_challenger_summonerId_list.append(summonerId)
    time.sleep(0.2)
    
    print(f'{year}년 {month}월 {day}일 {hour}시 기준\n{user_name}님은 {tier} {user_rank}입니다.\n{user_wins}승, {user_losses}패 기록 중 입니다.')
    print(f'summonerID = {summonerId}')
    print()    
```

```python
kr_challenger_df = pd.DataFrame({'user_name' : kr_challenger_username_list,
                                'user_rank' : kr_challenger_userrank_list,
                                'user_wins' : kr_challenger_userwins_list,
                                'user_losses' : kr_challenger_userlosses_list,
                                'summonerId' : kr_challenger_summonerId_list
                                })

kr_challenger_df
```

```python
time.sleep(10)
kr_challenger_accountId = []
kr_challenger_puuid = []
kr_challenger_user_name = []

for summonerId in tqdm(kr_challenger_df['summonerId']):
    puuid_request_url = f'https://kr.api.riotgames.com/tft/summoner/v1/summoners/{summonerId}?api_key{my_key}'
    puuid_res = urllib.request.Request(puuid_request_url)

    for key, value in headers.items():
        puuid_res.add_header(key,value)

    response = urllib.request.urlopen(puuid_res)
    result = response.readlines()

    result = json.loads(result[0].decode('utf-8'))

    if response.getcode() != 200:
        print('Error')

    user_name = result.get('name')
    accountId = result.get('accountId')
    puuid = result.get('puuid')
    
    kr_challenger_user_name.append(user_name)
    kr_challenger_accountId.append(accountId)
    kr_challenger_puuid.append(puuid)
    
    time.sleep(0.5)
    
```

```python
temp = pd.DataFrame({'user_name': kr_challenger_user_name,
                     'accountId' : kr_challenger_accountId,
                     'puuid' : kr_challenger_puuid
                    })

kr_challenger_all_info = pd.merge(kr_challenger_df, temp, how='right')
kr_challenger_all_info
```

```python
puuid_list = []
match_id = []

for puuid in tqdm(kr_challenger_all_info['puuid']):
    match_info_request_url = f'https://asia.api.riotgames.com/tft/match/v1/matches/by-puuid/{puuid}/ids?start=0&count=20&api_key={my_key}'
    match_info_res = urllib.request.Request(match_info_request_url)

    for key, value in headers.items():
        match_info_res.add_header(key,value)

    response = urllib.request.urlopen(match_info_res)
    result = response.readlines()

    result = json.loads(result[0].decode('utf-8'))

    if response.getcode() != 200:
        print('Error')
    
    match_info = result
    puuid_list.append(puuid)
    match_id.append(match_info)
    time.sleep(0.2)
```

```python
temp = pd.DataFrame({'puuid': puuid_list,
                     'match_id' : match_id
                    })

kr_challenger_all_info = pd.merge(kr_challenger_all_info, temp, on='puuid', how='right')
kr_challenger_all_info
```

```python
match_id_list = []
user_name_list = []
trait_name_list = []
for user_name, puuid, player in zip(kr_challenger_all_info['user_name'], kr_challenger_all_info['puuid'], kr_challenger_all_info['match_id']):
    for match_id in player:
        match_id_request_url = f'https://asia.api.riotgames.com/tft/match/v1/matches/{match_id}?api_key={my_key}'
        match_res = urllib.request.Request(match_id_request_url)

        for key, value in headers.items():
            match_res.add_header(key,value)

        response = urllib.request.urlopen(match_res)
        result = response.readlines()

        result = json.loads(result[0].decode('utf-8'))

        if response.getcode() != 200:
            print('Error')

        for match in tqdm(result.get('info')):
            ignore_match = ['game_datetime', 'game_length', 'game_version', 'queue_id', 'tft_game_type', 'tft_set_core_name', 'tft_set_number']
            if match in ignore_match:
                continue
            
            for participant in result.get('info').get(match):
                match_puuid = participant.get('puuid')
                if match_puuid != puuid:
                    continue
                else:
                    match_id_list.append(match_id)
                    user_name_list.append(user_name)
                    
                    trait_name = []
                    trait_style_list = []
                    for traits in participant.get('traits'):
                        trait_1_name = traits.get('name')
                        trait_style = traits.get('style')
                        trait_name.append(trait_1_name)
                        trait_style_list.append(trait_style)
                        
#                         time.sleep(0.2)
                    
                    trait_name_list.append([trait_name, trait_style_list])
                    
    break
                
```

```python
chal_info = pd.DataFrame({'match_id' : match_id_list,
                          'user_name': user_name_list,
                          'traits' : trait_name_list
                         })

chal_info
```

```python
match_id_list = []
user_name_list = []
augment_total = []

for user_name, puuid, player in zip(kr_challenger_all_info['user_name'], kr_challenger_all_info['puuid'], kr_challenger_all_info['match_id']):
    for match_id in tqdm(player):
        match_id_request_url = f'https://asia.api.riotgames.com/tft/match/v1/matches/{match_id}?api_key={my_key}'
        match_res = urllib.request.Request(match_id_request_url)

        for key, value in headers.items():
            match_res.add_header(key,value)

        response = urllib.request.urlopen(match_res)
        result = response.readlines()

        result = json.loads(result[0].decode('utf-8'))

        if response.getcode() != 200:
            print('Error')

        for match in result.get('info'):
            ignore_match = ['game_datetime', 'game_length', 'game_version', 'queue_id', 'tft_game_type', 'tft_set_core_name', 'tft_set_number']
            if match in ignore_match:
                continue
                
            for participant in result.get('info').get(match):
                match_puuid = participant.get('puuid')
                if match_puuid != puuid:
                    continue
                else:
                    match_id_list.append(match_id)
                    user_name_list.append(user_name)
                    
                    augment_list = []
                    for augments in participant.get('augments'):
                        augment_list.append(augments)
                    augment_total.append(augment_list)
                    
                    time.sleep(0.2)
    break
```

```python
print(len(match_id_list))
print(len(user_name_list))
print(len(augment_total))
```

```python
temp = pd.DataFrame({'match_id' : match_id_list,
                     'user_name': user_name_list,
                     'augments' : augment_total
                    })

chal_info = pd.merge(chal_info, temp, on=['match_id', 'user_name'], how='right')
chal_info
```
