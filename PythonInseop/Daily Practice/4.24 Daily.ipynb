```python
# 연습문제 1: 와인 데이터셋을 불러와 데이터프레임으로 만들어보세요.
wines = pd.read_csv("winequality.csv")
```

```python
# 연습문제 2: 기술통계 정보를 출력해보세요.
wines.info()
wines.describe ()
```

```python
# 연습문제 3: 와인의 종류(quality)별로 데이터가 몇 개씩 있는지 출력해보세요.

wines["quality"].value_counts()
```

```python
# 연습문제 7: 각 와인의 타입(type)별로 평균 알코올 도수(alcohol)와 평균 당도(residual sugar)를 구해보세요.
# print(wines["type"])
print(wines.groupby("type")["alcohol"].mean())
print()
print(wines.groupby("type")["residual sugar"].mean())
```

```python
# 연습문제 8: 'quality' 열의 값이 5 이상인 데이터만 추출하시오.
wines_qulaity = wines[wines["quality"] >= 5]
wines_qulaity
wines_qulaity.sort_values(by = "quality", ascending = False)
```

```python
# 연습문제 9: 'fixed acidity' 열과 'pH' 열의 값이 각각 3 이상, 4이상 인 데이터만 추출하여 데이터 프레임을 생성하시오.
wines[(wines["fixed acidity"]>= 3) | (wines["pH"] >= 4)]
```

```python
# 연습문제 10: 'volatile acidity' 열의 값이 0.5 이하인 데이터의 'quality' 열의 평균을 계산하세요.
low_volatile = wines[wines["volatile acidity"] <= 0.5]
low_volatile["quality"].mean()
```

```python
# 연습문제 11: 'residual sugar' 열의 값이 2 이하인 데이터의 'density' 열의 중앙값을 계산하세요.
low_residual = wines[wines["residual sugar"] <= 2]
low_residual["density"].median()
```

```python
# 연습문제 12: 'alcohol' 열이 가장 높은 5개 데이터의 'quality' 열 값을 출력하세요.
wines.sort_values(by=["alcohol"], ascending=False).head(5)[["quality", "alcohol"]]
```

```python
# 연습문제 13: 'chlorides' 열의 값이 가장 높은 데이터 10개의 'pH' 열 값을 출력하세요.
wines.sort_values(by=["chlorides"], ascending=False).head(10)[["pH", "chlorides"]]
```

```python
# 연습문제 14: 'sulphates' 열의 값이 0.4 이하인 데이터의 'quality' 열 값을 1로 변경하세요.
qual_low_sul = wines.loc[wines["sulphates"] <= 0.4, 'quality'] = 1
wines[wines["sulphates"] <= 0.4]
```

```python
# 연습문제 15: 'fixed acidity' 열의 값이 7 이하인 데이터의 'quality' 열 값을 0으로, 8 이상인 데이터의 'quality' 열 값을 2로 변경하세요.
# 나머지 데이터는 변경하지 않습니다.

low_ac = wines.loc[wines["fixed acidity"] <= 7, "quality"] = 0
high_ac = wines.loc[wines["fixed acidity"] >= 8, "quality"] = 2
```

```python
# wines.loc[(wines["fixed acidity"] <= 7) & (wines["fixed acidity"] >= 8), "quality"] = wines.loc[(wines["fixed acidity"] <= 7) & (wines["fixed acidity"] >= 8), "quality"].map({0: 0, 2: 2})
# wines[wines["fixed acidity"] <= 7]["quality"]

concat_data_df = pd.concat([low_ac, high_ac], ignore_index=True)
concat_data_df["quality"].value_counts()
```

```python
wines[wines["fixed acidity"] >=8]["quality"]
```

```python
# 연습문제 16: 'quality' 열의 값이 0인 데이터 중 'fixed acidity' 열의 값을 7로 변경하세요.
wines.loc[wines["quality"] == 0, "fixed acidity"] = 7
print(wines[wines["quality"] == 0]["fixed acidity"])

print()

zero_quality_df = concat_data_df[concat_data_df["quality"]==0]
zero_quality_df.loc[:,"fixed acidity"] = 7
print(zero_quality_df)
```

```python
# 연습문제 17: 'quality' 열의 값이 6 이하인 데이터 중 'pH' 열의 값을 3으로 변경하세요.
wines.loc[wines["quality"] <= 6, "pH"] = 3
wines[wines["quality"] <= 6]["pH"]
```
