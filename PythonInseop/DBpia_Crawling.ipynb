```python
import selenium
import os
import chromedriver_autoinstaller
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

options = webdriver.ChromeOptions()

# options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
browser = webdriver.Chrome("chromedriver", options=options)

# ------------------------------------------------시작------------------------------------------------------------

page = 1
# n = 3
thesis_data_list = []

        
url = "https://www.dbpia.co.kr/search/topSearch?searchOption=all&query=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5"
browser.get(url)

while page < 4:
    time.sleep(2)
    select_thesis = browser.find_elements(By.CLASS_NAME, "thesis__tit")
    print(f"Page {page}")

    time.sleep(2)
    
    select_thesis = browser.find_elements(By.CLASS_NAME, "thesis__tit")
    
# ------------------------------------------------탭 전환----------------------------------------------------------

    for thesis in select_thesis:
        thesis.click()
        last_tab = browser.window_handles[-1]
        browser.switch_to.window(window_name=last_tab)
        
# -------------------------------------------------논문 시작--------------------------------------------------------    
# ------------------------------------------------제목 가져오기------------------------------------------------------    


        thesis_title = browser.find_element(By.CLASS_NAME, "thesisDetail__tit").text

# ------------------------------------------------저자 가져오기------------------------------------------------------

        thesis_authors = browser.find_elements(By.CSS_SELECTOR, ".thesisDetail__author > .authorList > li")
        thesis_author_list = []
        for thesis_author in thesis_authors:
            thesis_author_list.append(thesis_author.text)
        thesis_author_string = ", ".join(thesis_author_list)
        thesis_author_array = ["", "", "", "", ""]
        minLength = min([len(thesis_author_array),len(thesis_author_list)])
        for i in range(minLength):
            thesis_author_array[i] = thesis_author_list[i]
    
    
    
# ----------------------------------------------이용 수 가져오기-----------------------------------------------------

        thesis_ref_counts = browser.find_element(By.ID, "showUsageChartButton").text

# ------------------------------------------------초록 가져오기------------------------------------------------------


        thesis_intro = browser.find_elements(By.CLASS_NAME, "abstractTxt");
        intro_word = ""
        if len(thesis_intro) > 0:
            intro_word = thesis_intro[0].text
# ------------------------------------------------논문 한개 끝------------------------------------------------------
# ---------------------------------------------Dictionary 만들기---------------------------------------------------

        thesis_data = {"Thesis Title": thesis_title,
                       "Thesis Usage": thesis_ref_counts,
                       "Thesis Intro": intro_word,
                       "Thesis Author(s)": thesis_author_string}

# ----------------------------------------------리스트에 Append-----------------------------------------------------

        thesis_data_list.append(thesis_data)

# -----------------------------------------------현재 탭 닫기--------------------------------------------------------

        browser.close()
        first_tab = browser.window_handles[0]
        browser.switch_to.window(window_name=first_tab)

# ------------------------------------------다시 위로 반복 / 다음 논문--------------------------------------------------
    
    try: 
        browser.find_element(By.CLASS_NAME, "dpPaging").find_elements(By.TAG_NAME,"a")[page].click()
        page += 1
        time.sleep(2)
    
    except IndexError:
            
        click = browser.find_element(By.ID, "goNextPage")
        click.send_keys("\n")        
#             click.click()
        page += 1
        time.sleep(2)

            

# # ------------------------------------------다시 위로 반복 / 다음 페이지-------------------------------------------------

thesis_data_df = pd.DataFrame(thesis_data_list)
thesis_data_df.to_csv("./DBPia_Crawling_Final.csv", encoding="utf-8-sig")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
import time
import pandas as pd
from pandas import DataFrame


options = webdriver.ChromeOptions()

# options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
driver = webdriver.Chrome("chromedriver", options=options)

# # 여기서 부터 ---->
# service = Service("/usr/bin/chromedriver")

# options = Options()
# # options.add_argument("--headless")
# options.add_argument("--no-sandbox")


# driver = webdriver.Chrome(service=service, options=options)
# # 화면이 작아지면 보이는 것과 다르니 본인이 사용하는 해상도를 꼭 맞추세요!!
# driver.set_window_size(1120, 1150)
# # <---- 여기까지는 본인이 쓰던 데로 쓰세요

# 디비 피아 논문 인공지능 검색
driver.get("https://www.dbpia.co.kr/search/topSearch?searchOption=all&query=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5#a")
time.sleep(1)

# 논문 링크 1~3페이지 수집용
paperUrlList = []

# -----------------------------------------------------------------
# 1 페이지 이동
driver.find_elements(By.CSS_SELECTOR, "#pageList > a")[0].click()
# 페이지 이동 후 화면처리가 존재하므로 딜레이 필요 안되면 초를 늘리세요.
time.sleep(1)
# 1페이지 수집
# class이름으로 논문 링크 객체 수집
paperList = driver.find_elements(By.CLASS_NAME, "thesis__link")

# 논문 링크 객체에서 URL 수집
for paper in paperList:
    paperUrlList.append(paper.get_attribute("href"))

# -----------------------------------------------------------------
# 2 페이지 이동
driver.find_elements(By.CSS_SELECTOR, "#pageList > a")[1].click()
# 페이지 이동 후 화면처리가 존재하므로 딜레이 필요 안되면 초를 늘리세요.
time.sleep(1)

# 2페이지 수집
# class이름으로 논문 링크 객체 수집
paperList = driver.find_elements(By.CLASS_NAME, "thesis__link")

# 논문 링크 객체에서 URL 수집
for paper in paperList:
    paperUrlList.append(paper.get_attribute("href"))

paging = driver.find_elements(By.CSS_SELECTOR, "#pageList > a")


# -----------------------------------------------------------------
# 3 페이지 이동
driver.find_elements(By.CSS_SELECTOR, "#pageList > a")[2].click()
# 페이지 이동 후 화면처리가 존재하므로 딜레이 필요 안되면 초를 늘리세요.
time.sleep(1)

# 3페이지 수집
# class이름으로 논문 링크 객체 수집
paperList = driver.find_elements(By.CLASS_NAME, "thesis__link")

# 논문 링크 객체에서 URL 수집
for paper in paperList:
    paperUrlList.append(paper.get_attribute("href"))

paging = driver.find_elements(By.CSS_SELECTOR, "#pageList > a")


# 논문 정보 저장용
paperInfoList = []

# 논문 링크 마다 데이터 수집
# [:3] 테스트를 위해 3개만 먼저 수행 다 도릴 때는 삭제
for paperUrl in paperUrlList[:3]:

    # 논문 접속
    driver.get(paperUrl)

    # 타이틀 가져오기
    # ID 가 존재하면 ID로 단일 선택
    title = driver.find_element(By.ID, "thesisTitle").text

    # 이용수 가져오기
    # ID가 존재하니 ID로 단일 선택
    useCount = driver.find_element(By.ID, "showUsageChartButton").text

    # 키워드 가져오기
    # class 이름만 존재. class로 가져올 때는 가급적 다건으로 가져오자 find_element*s*
    # 다건 중 첫번째 것만 추출
    keywordElements = driver.find_elements(By.CLASS_NAME, "abstractTxt");
    # 키워드가 찾아지지 않으면 빈값으로 되게
    keyword = ""
    # 키워드가 찾아지면 입력
    if len(keywordElements) > 0:
        keyword = keywordElements[0].text

    # 저자정보 가져오기
    # 저자는 여러명 일 수 있으므로 css selector를 이용해서 다건으로 가져오자 find_element*s*
    authorElements = driver.find_elements(By.CSS_SELECTOR, ".thesisDetail__author > .authorList > li")
    
    # 가져온 객체에서 텍스트 추출해서 배열로 만들기
    authorList = []
    for author in authorElements:
        authorList.append(author.text)
    
    # 나중에 한칸의 셀에 , 로 구분해서 넣기용
    authorStrList = ",".join(authorList)

    # 저자 나누기
    authorArray = ["", "", "", "", ""]

    # 실제 저자가 5명 미만인 경우 오류 방지용
    minLength = min([len(authorArray), len(authorList)])

    # 저자가 5명 넘어도 제한해서 널 수 있게
    for i in range(minLength):
        authorArray[i] = authorList[i]

    # 딕션으로 만들기
    paperInfo = {
        "title" : title
        , "useCount" : useCount
        , "keyword" : keyword
        , "authorList" : authorList
        , "authorStrList" : authorStrList
        , "author1" : authorArray[0]
        , "author2" : authorArray[1]
        , "author3" : authorArray[2]
        , "author4" : authorArray[3]
        , "author5" : authorArray[4]
    }

    paperInfoList.append(paperInfo)

    # 1개 가져오고 1초 쉬기
    time.sleep(1)

# pandasd용 데이터 프레임으로 전환
data_df = DataFrame(paperInfoList)
data_df
# CSV 변환
# data_df.to_csv("파일이름.csv")
# 엑셀 변환
# data_df.to_excel("파일이름.xls")
```
