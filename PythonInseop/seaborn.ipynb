```python
# 나이를 한번 구분해보겠습니다.

df["Age"].unique()

def set_age_category(age):
    category =""
    if age <= 5:
        category = "Children"
    elif age <= 18:
        category = "Student"
    elif age <= 30:
        category = "Young Adult"
    elif age <= 60:
        category = "Adult"
    else:
        category = "Elder"
    
    return category

df["Age_Category"] = df["Age"].apply(lambda x : set_age_category(x))
```

```python
df[["Age","Age_Category"]]
```

```python
sns.barplot(x="Age_Category",y="Survived", data=df, hue="Sex")
```

```python
sns.barplot(data=df, x='Age_Category', y='Survived', hue='Sex', 
            order=['Children', 'Student', 'Young Adult', 'Adult', 'Elder'])
```

```python
orders=['Children', 'Student', 'Young Adult', 'Adult', 'Elder']

sns.barplot(data=df, x="Age_Category",y="Survived", hue="Sex", order=orders)
```

```python
# 남성과 여성별로 나이대의 생존확률 분포도를 나타내시오

orders=['Children', 'Student', 'Young Adult', 'Adult', 'Elder']

sns.barplot(data=df, x="Sex",y="Survived", hue="Age_Category")
```

### subplots을 사용해서 여러개의 그래프 한번에 표현하기

```python
columns = ["Survived","Pclass","Sex","Age_Category"]
for index, data in enumerate(columns):

    print(index, data)
```

```python
columns = ["Survived","Pclass","Sex","Age_Category"]
fig, ax = plt.subplots(1,4, figsize = (16,8))

for index, column in enumerate(columns):
    sns.barplot(x=column, y="Survived", data=df, ax=ax[index])
```
