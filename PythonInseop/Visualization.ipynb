```python
x_value01 = np.arange(1,10)
x_value02 = np.arange(1,20)

y_value01 = 2*x_value01
y_value02 = 2*x_value02

fig, (ax_01, ax_02) = plt.subplots(nrows=1, ncols=2, figsize=(12,6))

ax_01.plot(x_value01, y_value01, color="red", marker = "o", markersize = 6, linestyle="dashed", linewidth=2, label = "line")
ax_02.bar(x_value02, y_value02, color="blue", label = "bar")

ax_01.set_xlabel("ax_01 x label")
ax_02.set_xlabel("ax_02 x label")

ax_01.legend()
ax_02.legend()

plt.show()
```

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 6)) # subplot은 하나씩 설정, subplots은 한번에 설정.

# 첫번째 그래프
ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6, label='temp_01')

# 두번째 그래프
ax[0][1].bar(x_value_02, y_value_02, color='green', label='temp_02')

# 세번째 그래프
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=6, label='temp_03')

# 네번재 그래프
ax[1][1].bar(x_value_02, y_value_02, color='red', label='temp_04')

ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

ax[0][0].legend()
ax[0][1].legend() 
ax[1][0].legend()
ax[1][1].legend() 

#plt.legend()
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/datagy/data/main/KNN_data.csv')
df.head()
df.plot()
```

```python
# Label => Large, Medium, Small 각각 df를 만드시오.

small_df = df[df["Label"] == "Small"]
medium_df = df[df["Label"] == "Medium"]
large_df = df[df["Label"] == "Large"]

ax = small_df.plot(x='x', y='y', kind= 'scatter', color = "r", label= "small")
medium_df.plot(x='x', y='y', kind= 'scatter', color = "b", label= "medium", ax = ax)
large_df.plot(x='x', y='y', kind= 'scatter', color = "g", label= "large",ax=ax)
plt.scatter(medium_df["x"], medium_df["y"], color= "violet", marker="v", label = "Medium_violet")
plt.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "AppleGothic"
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

data = {
    '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
## 1. 영화 데이터를 활용하여 x 축은 영화, y축은 평점인 막대 그래프를 만드시오.
```

```python
plt.xticks(fontsize=10, rotation=45)
ax = plt.bar(df["영화"], df["평점"], alpha = 0.5, width = 0.5)
```

```python
## 2. 앞에서 만든 막대 그래프에 제시된 세부 사항을 적용하시오.
#- 제목: 국내 Top 8 영화 평점 정보
#- x축 label: 영화 (90도 회전)
#- y축 label: 평점

plt.xticks(fontsize=10, rotation=45)
ax = plt.bar(df["영화"], df["평점"], alpha = 0.5, width = 0.5)

plt.title("국내 Top 8 영화 평점 정보")
plt.xlabel('영화')
plt.xticks(rotation=25)

plt.ylabel('평점')



plt.show()
```

```python
## 3. 개봉 연도별 평점 변화 추이를 꺾은선 그래프로 그리시오.
#연도별 평균 데이터를 구하는 코드는 다음과 같습니다.
df_year_rating = df[['개봉 연도', '평점']]
df_group = df_year_rating.groupby('개봉 연도').mean()
plt.figure(figsize=(10, 6))

plt.plot(df_group.index, df_group['평점'], marker='o', linestyle='-', color='b')
plt.xlabel('개봉 연도')
plt.ylabel('평균 평점')
plt.title('개봉 연도별 평균 평점 변화 추이')
plt.grid()
plt.show()
```

```python
df_group= df.groupby("개봉 연도").mean(numeric_only=True)
movie_graph = plt.plot(df_group.index, df_group["평점"])
plt.show()
```

```python
## 4) 앞에서 만든 그래프에 제시된 세부 사항을 적용하시오.
#- marker: 'o'
#- x축 눈금: 5년 단위 (2005, 2010, 2015, 2020)
#- y축 범위: 최소 7, 최대 10

movie_graph = plt.plot(df_group.index, df_group["평점"], marker ="o")
plt.xticks([2005, 2010, 2015, 2020])
plt.ylim(7,10)

plt.show()
```

```python
## 5) 평점이 9점 이상인 영화의 비율을 확인할 수 있는 원 그래프를 제시된 세부 사항을 적용하여 그리시오.
#- label: 9점 이상 / 9점 미만
#- 퍼센트: 소수점 첫 번째 자리까지 표시
#- 범례: 그래프 우측에 표시
nineorabove = df[df["평점"] >= 9.0]
belownine = df[df["평점"] < 9.0]
# nineorabove
# belownine

values = [len(nineorabove), len(belownine)]
determinant = ["평점 9점 이상", "평점 9점 미만"]

plt.pie(values, labels=determinant, autopct="%.1f")
plt.legend(loc=(1,0.3))
plt.show()
```
