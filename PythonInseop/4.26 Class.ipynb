```python
# 라이브러리 import
import os
import numpy as np
import pandas as pd
import folium
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 데이터 파일 read
seoul_file = pd.read_excel("./xlsx etc/seoul_popular_data.xls", header=2, usecols="B,D,G,J,N")
```

```python

# 컬럼명 변경:
# 자치구 -> 구, 합계 -> 인구수, 등록외국인 - > 외국인, 65세이상고령자 -> 고령자
seoul_file.columns
seoul_file.columns=["구","인구수","한국인","외국인","고령자"]
```

```python
# 함계행 제거
seoul_file.drop([0],inplace=True)
```

```python
new_seoul_file = seoul_file.reset_index()
new_seoul_file
```

```python
new_seoul_file.drop('index',inplace=True,axis=1)
new_seoul_file
```

```python
## (4) 각 구별 전체 인구를 이용해서 구별 "외국인비율"과 "고령자비율"을 추가하고, 소수점 첫째자리 반올림하기
### - 외국인 비율: 외국인/인구수 * 100
### - 고령자 비율: 고령자 / 인구수 * 100

new_seoul_file["인구수"]= new_seoul_file["인구수"].astype("int")
new_seoul_file["한국인"] = new_seoul_file["한국인"].astype("int")
new_seoul_file["외국인"] = new_seoul_file["외국인"].astype("int")
new_seoul_file["고령자"] = new_seoul_file["고령자"].astype("int")


new_seoul_file["외국인비율"] = round((new_seoul_file["외국인"]/new_seoul_file["인구수"])*100, 1)
new_seoul_file["고령자비율"] = round((new_seoul_file["고령자"]/new_seoul_file["인구수"])*100, 1)

new_seoul_file
```

```python
# (5) 인구수별로 내림차순 정렬(sort)하기
new_seoul_file.sort_values(by="인구수", ascending=False)
```
