```python
# Pclass 별 탑승객 수를 구하시오
titanic_data["Pclass"].value_counts()
```

```python
# Pclass 별 Embarked (정착장) 수를 계산하시오
# titanic_data["Pclass"].groupby("Embarked")
titanic_data[["Pclass","Embarked"]].value_counts()
```

```python
# SibSP(함께 탑승한 형제자매) + Parch (함께 탑승한 부모자녀)
# => 전체가족이 많으면 => 생존률이 높을까? 혼자가 생존률이 높을까?

titanic_data["Total_Family_Num"] = titanic_data["SibSp"] + titanic_data["Parch"] + 1
print(titanic_data.info())
titanic_data
```

```python
# EDA => 데이터 훑는 과정

titanic_fare = titanic_data["Fare"]
print(titanic_fare.max())
print(titanic_fare.min())
print(titanic_fare.mean())
print(titanic_fare.sum())
```

```python
titanic_data[titanic_data["Age"] >= 60][["Name", "Age", "Survived", "Sex"]]
```

```python
# 나이가 비어있는 데이터들이 있습니다. 비어있는 데이터에 나이의 평균값을 입력하는 코드를 작성하시오.

titanic_data["Age"].fillna(titanic_data["Age"].mean())
df.isnull().sum()
```
