```python
from collections import defaultdict

def cal_days():
  year, month, day = map(int, today.split(".")) # [2021, 05, 02]

  # day_list = today.split(".") #[2021, 05 , 02]
  # 2000년 1월 1일을 기준으로 일 수를 계산
  return (year - 2000) * 12 * 28 + month * 28 + day

# today -> day로 변경
def solution(today, terms, privacies):
  
  today = cal_days(today)

  terms_dict = defaultdict()

  for term in terms:
    alp, month = term.split()
    terms_dict[alp] = int(month) * 28 # {"a" : 6}

  answer = []

  for i in range(len(privacies)):
    collected_day, alp = privacies[i].aplit()

    if today >= terms_dict[alp] + cal_days(collected_day):
      answer.append(i+1)

  return answer
```
