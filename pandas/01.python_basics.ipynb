## 조건문 - if
1. if 문이란
if의 뜻, 만약에~ => 가 참이면 실행 / 거짓이면 다음 조건문 실행

```python
number = int(input("숫자를 입력하세요 : "))

if number < 5:
  print("숫자가 5보다 작습니다")

elif number < 10:
  print("숫자가 10보다 작습니다.")

else:
  print("10보다 큽니다")
```

```python
# 서울에서 부산을 간다고 가정하에 나의 주머니에 있는 돈(nmoney)에 따른 조건문

# if => 7만원이 있으면 비행기를 타고
# elif => 5만원이 있으면 기차를 타고
# elif => 3만원이 있으면 버스를 타고
# else => 걸어간다

money = int(input("주머니에 얼마 있습니까? "))

if money == 70000:
  print("비행기")
elif money == 50000:
  print("기차")
elif money == 30000:
  print("버스")
else:
  print("걸어간다")
```

* 조건문이 참인 경우 문장이 반복되서 실행

```python
# 1부터 5까지 더하는 프로그램을 만들어보시오.
# for, while

sum = 0

for i in range(1, 6):
  print(i)
  sum += i
  print()
print("for 최종값", sum)

print()

sum = 0
i = 0

while i < 5:
  i += 1
  sum += i


print("While 최종값", sum)    
```

```python
print("break")
# 반복문 제어문 - continue, break

for i in range(10):
  if 3 <= i < 5: # 3, 4
    break

  print (i)

print()
print("continue")
# 반복문 제어문 - continue, break

for i in range(10):
  if 3 <= i < 5: # 3, 4
    continue

  print (i)
```

```python
#q. 성적과 이름을 입력하고 (매개변수), 성적에 따라 이름과 등급을 알려주는 프로그램 (함수)를 만드시오.

# 90점 이상 A등급
# 80점 이상 B등급
# 70점 이상 C등급
# 60점 이상 D등급
# 60점 미만 F등급

def rank(score, name):
  if score >= 90:
    rank = "A"
  elif score >= 80:
    rank = "B"
  elif score >= 70:
    rank = "C"
  elif score >= 60:
    rank = "D"
  else:
    rank = "F"
  print(f"Hi {name}, your score is {score}, your rank is {rank}")

score, name = rank(int(input("please enter your score: ")), input("please enter your name: "))

```

## 3. 클래스 Class

  * 함수가 특정기능을 수행하는 일련의 코드블록을 나타내는데, 클래스는 함수보다는 더 큰 개념. 함수가 여러개 모여 있는 것

Class랑 def가 뭐가 달라요?

```python
class FishBread():
  def redbean_bread(self):
    print("팥 붕어빵")

  def chouxcream_bread(self):
    print("슈크림 붕어빵")

a = FishBread() # class

a.redbean_bread()
```

```python
class FishBread:
 # 생성자 => 클래스가 호출될 때 가장 먼저 실행되는 메소드
 def __init__(self, name, ingredient, price):
  self.name = name
  self.ingredient = ingredient
  self.price = price

 def bread(self):
  print("이 붕어빵의 이름은 {}이고, {}이 들어가있습니다. 가격은 {} 입니다.".format(self.name, self.ingredient, self.price))


redbean_bread = FishBread("팥붕어빵", "팥", 500)
redbean_bread.bread()

print(redbean_bread.name)
print(redbean_bread.ingredient)
print(str(redbean_bread.price))

cream_bread = FishBread("슈크림 붕어빵", "크림", 700)
cream_bread.bread()

pizza_bread = FishBread("피자 붕어빵", "피자", 1000)
pizza_bread.bread()
```
